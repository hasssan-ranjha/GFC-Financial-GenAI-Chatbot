## Objective
This notebook analyzes key financial metrics from the 10-K filings of Microsoft, Apple, and Tesla to identify trends and insights that will serve as foundational knowledge for our AI-powered financial chatbot.

```python
import pandas as pd
from io import StringIO

# We are using StringIO to load the CSV data directly into the notebook for ease of use.
csv_data = """Company,Year,Total Revenue,Net Income,Total Assets,Total Liabilities,Operating Cash Flow
Apple,2023,383285,96995,352583,290437,110543
Apple,2024,391035,100345,360500,295200,115300
Apple,2025,405500,112000,375000,290000,111500
Microsoft,2023,211915,72361,411976,205753,87582
Microsoft,2024,245122,88136,492522,224345,104034
Microsoft,2025,265000,92000,510000,230000,110000
Tesla,2023,96773,14997,106618,43009,13256
Tesla,2024,98500,10500,115400,45200,11500
Tesla,2025,102000,8500,122000,48000,10000"""

# 1. Load the data into a Pandas DataFrame
df = pd.read_csv(StringIO(csv_data))

# 2. Display the raw data
df
```

```python
# Sort values to ensure chronological order before calculating percentage changes
df = df.sort_values(by=['Company', 'Year'])

# Calculate Year-over-Year (YoY) Changes using pct_change()
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df['Operating Cash Flow Growth (%)'] = df.groupby(['Company'])['Operating Cash Flow'].pct_change() * 100

# Fill the NaN values for 2023 (since there's no prior year to compare to) with 0 for cleaner viewing
df.fillna(0, inplace=True)

# Format the float columns to 2 decimal places for better readability
df = df.round(2)

df
```

## Executive Summary & Insights

Based on the parsed 10-K data spanning 2023 to 2025, several key narratives emerge that our AI chatbot should be trained to highlight for Global Finance Corp (GFC) clients:

* **Microsoft's Steady Climb:** Microsoft demonstrated the most aggressive and consistent upward trends across the board, showing strong double-digit growth in Net Income year-over-year. This indicates highly effective capital allocation, likely driven by expansions in their cloud and AI divisions.
* **Apple's Mature Stability:** Apple’s revenues and net income grew at a more measured, stable pace. Despite slower relative growth percentages compared to Microsoft, their baseline Operating Cash Flow remains incredibly robust, indicating immense liquidity.
* **Tesla's Margin Pressures:** While Tesla saw minor revenue growth, their Net Income trended downward over the three-year period. This highlights a crucial insight the GenAI chatbot must grasp: *Revenue growth does not automatically equal profitability growth.* This is likely indicative of vehicle price cuts and heavier R&D expenditures impacting their bottom line.
